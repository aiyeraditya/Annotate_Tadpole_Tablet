```python
import os
import cv2
import pandas as pd
import numpy as np
```

```python
def get_boundary(x, b = 2048, margin = 300):
    x_mean = np.nanmean(x[:6])
    if x_mean < margin/2:
        return 0, margin
    if x_mean > b - margin/2:
        return b-margin, b
    return int(x_mean - margin/2), int(x_mean + margin/2)
```

```python
def crop_folder(path_):
    df_ = pd.read_hdf(f'{path_}CollectedData_Aditya.h5')
    cols_ = df_.index
    for c in cols_[1:]:
        A = np.array(df_.loc[c])
        x = A[::2]
        y = A[1::2]
        x_range = get_boundary(x, 2048, 575)
        y_range = get_boundary(y, 1536, 575)
        I = cv2.imread(path_ + c[2])
        A[::2] = x - x_range[0]
        A[1::2] = y - y_range[0]
        df_.loc[c] = A
        I = I[y_range[0]:y_range[1], x_range[0]:x_range[1]]
        cv2.imwrite(path_ + c[2], I)
    df_.to_hdf(f'{path_}CollectedData_Aditya.h5', 'df_with_missing')
    df_.to_csv(f'{path_}CollectedData_Aditya.csv')
```

```python
p_ = '/home/iyer_la/Documents/tadpole_stage57_Cam0-Aditya-2022-11-04/labeled-data/';
folders = [p + i + '/' for i in os.listdir(p) if 'labeled' not in i]
for i in folders:
    crop_folder(i)
```
